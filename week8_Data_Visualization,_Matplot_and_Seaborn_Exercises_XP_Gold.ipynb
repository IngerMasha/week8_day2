Exercise 1 : Bar Graph
Instructions
Create a bar graph that represents the sales data of a retail company, where the x-axis represents different product categories, and the y-axis represents the total sales amount. Use a fictional dataset for this exercise.


```python
import matplotlib.pyplot as plt

categories = ['Cars', 'Clothing', 'Food', 'Jewelry', 'Toys']
sales_amount = [20000, 15000, 30000, 10000, 12000]

plt.figure(figsize=(10, 6))
plt.bar(categories, sales_amount, color='skyblue')
plt.title('Total Sales Amount by Product Category')
plt.xlabel('Product Categories')
plt.ylabel('Sales Amount ($)')
plt.show()
```

Exercise 2 : Hierarchical Indexing

```python
import pandas as pd
import numpy as np

data = {
    'country': ['Canada', 'Canada', 'Canada', 'Canada', 'USA', 'USA', 'USA', 'USA'],
    'city': ['Toronto', 'Toronto', 'Vancouver', 'Vancouver', 'New York', 'New York', 'San Francisco', 'San Francisco'],
    'date': ['2024-07-01', '2024-07-02', '2024-07-01', '2024-07-02', '2024-07-01', '2024-07-02', '2024-07-01', '2024-07-02'],
    'temperature': [28, 32, 29, 31, 25, 27, 22, 24]
}
df = pd.DataFrame(data)
df['date'] = pd.to_datetime(df['date'])

df.set_index(['country', 'city', 'date'], inplace=True)

print("Hierarchical DataFrame:")
print(df)
filtered_data = df.loc[('Canada', slice(None), slice(None))]
filtered_data = filtered_data[filtered_data['temperature'] > 30]
print("\nDays when the temperature exceeded 30°C in Canadian cities:")
print(filtered_data)
```

Exercise 3 : Advanced Filtering With Hierarchical Indices

```python
import pandas as pd

data = {
    'Department': ['HR', 'HR', 'IT', 'IT', 'Finance', 'Finance'],
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank'],
    'Salary': [48000, 52000, 58000, 45000, 60000, 55000]
}
df = pd.DataFrame(data)
df.set_index(['Department', 'Employee'], inplace=True)

print("Original DataFrame:")
print(df)
```

```python
filtered_df = df[df['Salary'] > 50000]

print("Rows where Salary is above 50,000 within each Department:")
print(filtered_df)
```

Exercise 5 : Data Visualisation Using MatPlotlib

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

x = np.linspace(0, 10, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y_sin, label='sin(x)', linestyle='-', color='b')
plt.plot(x, y_cos, label='cos(x)', linestyle='--', color='r')
plt.title('y = sin(x) and y = cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
products = ['Product A', 'Product B', 'Product C']
company_a = [20, 35, 30]
company_b = [25, 32, 34]

fig, ax = plt.subplots()
ax.bar(products, company_a, label='Company A')
ax.bar(products, company_b, bottom=company_a, label='Company B')
ax.set_xlabel('Products')
ax.set_ylabel('Sales')
ax.set_title('Stacked Bar Chart')
ax.legend()
plt.show()
```

```python
np.random.seed(42)
scores = np.random.randint(50, 100, 50)

plt.figure(figsize=(10, 6))
plt.hist(scores, bins=5, edgecolor='black')
plt.title('Histogram of Exam Scores')
plt.xlabel('Score')
plt.ylabel('Number of Students')
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

axs[0, 0].plot(x, y_sin, label='sin(x)', color='b')
axs[0, 0].set_title('Line Plot')
axs[0, 0].legend()

axs[0, 1].bar(products, company_a, color='g')
axs[0, 1].set_title('Bar Chart')

np.random.seed(42)
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)
axs[1, 0].scatter(x_scatter, y_scatter, color='m')
axs[1, 0].set_title('Scatter Plot')

axs[1, 1].hist(scores, bins=5, edgecolor='black', color='c')
axs[1, 1].set_title('Histogram')

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots()
plt.subplots_adjust(left=0.25, bottom=0.25)
scatter = ax.scatter(x_scatter, y_scatter, s=50)

ax_slider = plt.axes([0.25, 0.1, 0.65, 0.03])
slider = Slider(ax_slider, 'Size', 10, 200, valinit=50)

def update(val):
    scatter.set_sizes([slider.val] * len(x_scatter))
    fig.canvas.draw_idle()

slider.on_changed(update)

plt.show()
```

Exercise 6 : Data Visualisation

```python
import pandas as pd

file_path = '/content/sales_data.csv'
sales_data = pd.read_csv(file_path)
print("First few rows of the dataset:")
print(sales_data.head())
```

```python
total_revenue = sales_data['revenue'].sum()
print(f"Total revenue generated by the company over the entire year: ${total_revenue:.2f}")
```

```python
product_revenue = sales_data.groupby('product')['revenue'].sum()
highest_revenue_product = product_revenue.idxmax()
highest_revenue = product_revenue.max()
print(f"\nProduct '{highest_revenue_product}' generated the highest revenue of ${highest_revenue:.2f}")
```

```python
total_quantity = sales_data['quantity'].sum()
average_price_per_product = total_revenue / total_quantity
print(f"Average price of a product sold: ${average_price_per_product:.2f}")
```
