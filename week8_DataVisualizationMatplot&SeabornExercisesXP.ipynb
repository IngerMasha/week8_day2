Exercise 2 : Creating A Line Plot For Temperature Variation

```python
import matplotlib.pyplot as plt

days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
temperatures = [72, 74, 76, 80, 82, 78, 75]
plt.figure(figsize=(10, 6))
plt.plot(days, temperatures, marker='o', linestyle='-', color='b')
plt.xlabel("Day")
plt.ylabel("Temperature (Â°F)")
plt.title("Temperature Variation Over a Week")
plt.grid(True)
plt.show()
```

Exercise 3 : Visualizing Monthly Sales With A Bar Chart

```python
import matplotlib.pyplot as plt

months = ["January", "February", "March", "April", "May"]
sales = [5000, 5500, 6200, 7000, 7500]
plt.figure(figsize=(10, 6))
plt.bar(months, sales, color='pink')
plt.xlabel("Month")
plt.ylabel("Sales Amount ($)")
plt.title("Monthly Sales Data for a Retail Store")
plt.grid(axis='y')
plt.show()
```

 Exercise 3 : Visualizing Monthly Sales With A Bar Chart

```python
import matplotlib.pyplot as plt

monthly_sales = [5000, 5500, 6200, 7000, 7500, 8000, 8500, 9000, 9500, 10000, 10500, 11000]

months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

plt.figure(figsize=(10, 6))
plt.bar(months, monthly_sales, color='lightgreen')
plt.xlabel('Month')
plt.ylabel('Sales Amount ($)')
plt.title('Monthly Sales Data')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

Exercise 4 : Data Visualisation

```python
import pandas as pd
import matplotlib.pyplot as plt

sales_data = pd.read_csv('/content/sales_data.csv')
print(sales_data.head())
```

```python
total_quantity_sold = sales_data['quantity'].sum()
print(f"Total quantity of products sold: {total_quantity_sold}")

```

```python
sales_data['revenue'] = sales_data['quantity'] * sales_data['price']
category_revenue = sales_data.groupby('category')['revenue'].sum()
print(f"Total  revenue for each category: {category_revenue}")
```

```python
highest_revenue_category = category_revenue.idxmax()
highest_revenue = category_revenue.max()
print(f"Category with the highest revenue: {highest_revenue_category}")
print(f"Revenue generated by {highest_revenue_category}: ${highest_revenue:.2f}")
```

```python
average_revenue_per_sale = sales_data['revenue'].mean()
print(f"Average revenue per sale: ${average_revenue_per_sale:.2f}")
```

```python
sales_data['date'] = pd.to_datetime(sales_data['date'])
sales_data['quarter'] = sales_data['date'].dt.to_period('Q')

quarterly_revenue = sales_data.groupby('quarter')['revenue'].sum()

plt.figure(figsize=(10, 6))
quarterly_revenue.plot(kind='bar', color='skyblue')
plt.xlabel("Quarter")
plt.ylabel("Total Revenue ($)")
plt.title("Total Revenue Generated in Each Quarter")
plt.grid(axis='y')
plt.show()
```

Exercise 5 : Data Visualisation Using MatPlotLib

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 400)
y = x**2

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = x^2')
plt.title('Plot of y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
products = ['A', 'B', 'C', 'D']
sales = [15, 30, 45, 20]

plt.figure(figsize=(8, 6))
plt.bar(products, sales, color=['lightblue', 'lightgreen', 'pink', 'lightgrey'])
plt.title('Sales Data of Products A, B, C, and D')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.show()
```

```python
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [40, 30, 20, 10]
colors = ['red', 'yellow', 'pink', 'brown']

plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Favorite Fruits')
plt.legend(labels, loc="best")
plt.axis('equal')
plt.show()
```
